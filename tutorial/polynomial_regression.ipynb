```python
#Polynomial Regression

#if your data are not linear, you can use polynomial regression to fit your data
```

```python
#let's generate some nonlinear data
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
m = 100
X = 6 * np.random.rand(m, 1) - 3
Y = 0.5 * X**2 + X + 2 + np.random.rand(m, 1)

plt.scatter(X, Y, )
```

```python
#next, we create polynomial regressor
from sklearn.preprocessing import PolynomialFeatures
poly_feature = PolynomialFeatures(degree=2, include_bias=False)
x_poly = poly_feature.fit_transform(X)
print("original X[1]", X[1])
print("polynomial X[1]", x_poly[1])
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x_poly, Y)
lin_reg.intercept_, lin_reg.coef_
```

```python
y_pred_cal = 0.49967213 * X**2 + 0.9943591 * X + 2.49786712
y_pred = lin_reg.predict(x_poly)

#y_pred_cal are pretty much the same as y_pred
```

plt.plot(X, y_pred, alpha=0.2, color="R")
plt.scatter(X, Y, alpha=0.6)

```python
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(Y, y_pred))
```

### Another Example for Polynomial regression

```python
np.random.seed(0)
x = 2 - 3 * np.random.normal(0, 1, 20)
y = x - 2 * (x ** 2) + 0.5 * (x ** 3) + np.random.normal(-3, 3, 20)

plt.figure(figsize=(8, 7))
sns.scatterplot(x, y)
plt.show()
```

```python
lin = LinearRegression()
lin.fit(x.reshape(-1, 1), y.reshape(-1, 1))
y_pred_1d = lin.predict(x.reshape(-1, 1)) 

plt.figure(figsize=(8, 7))
sns.scatterplot(x, y)
sns.lineplot(x, y_pred_1d.flatten(), c='r')
plt.show()
```

### Add Polynomial features

```python
#degree 2
poly_feature = PolynomialFeatures(degree=2)
x_poly = poly_feature.fit_transform(x.reshape(-1, 1))
lin.fit(x_poly, y)
y_pred_2d = lin.predict(x_poly)

plt.figure(figsize=(8, 7))
sns.scatterplot(x, y)
sns.lineplot(x, y_pred_2d.flatten(), c='r')
plt.show()
```

```python
#degree 3
poly_feature = PolynomialFeatures(degree=3)
x_poly_d3 = poly_feature.fit_transform(x.reshape(-1, 1))
lin.fit(x_poly_d3, y)
y_pred_3d = lin.predict(x_poly_d3)

plt.figure(figsize=(8, 7))
sns.scatterplot(x, y)
sns.lineplot(x, y_pred_3d.flatten(), c='r')
plt.show()
```

```python
plt.figure(figsize=(8, 7))
sns.scatterplot(x, y)
sns.lineplot(x, y_pred_1d.flatten(), c='r', label='1degree')
sns.lineplot(x, y_pred_2d.flatten(), c='r')
sns.lineplot(x, y_pred_3d.flatten(), c='r')
plt.show()
```
