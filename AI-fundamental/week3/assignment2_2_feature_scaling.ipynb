```python
import numpy as np
import tensorflow as tf
import matplotlib
from matplotlib import pyplot as plt
import pandas as pd
from __future__ import division
```

```python
df = pd.read_excel('./data/fire_theft.xls')
```

```python
df.head()
```

```python
n_samples = df['X'].shape[0]
X = tf.placeholder(dtype='float32', name='X')
Y = tf.placeholder(dtype='float32', name='Y')

theta0 = tf.Variable(0.0, name='theta0')
theta1 = tf.Variable(0.0, name='theta1')
theta2 = tf.Variable(0.0, name='theta2')

y_pred = tf.multiply(tf.pow(X, 2), theta2) + tf.multiply(theta1,X) + theta0
loss_function = tf.divide(tf.reduce_sum(tf.pow((y_pred - Y), 2)), (2.0 * n_samples))

loss_summary = tf.summary.scalar(tensor=loss_function, name='loss_summary')

optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.03).minimize(loss_function)

def mean_normalize(xs):
    mean = xs.sum() / len(xs)
    std = np.sqrt(sum(((xs - mean)**2)) / (len(xs)))
    return (xs - mean) / std
    
with tf.Session() as sess:
    writer = tf.summary.FileWriter(logdir='./graphs/assignment_2_2_scaling/loss_summary')
    
    sess.run(tf.global_variables_initializer())
    
    for step in range(30000):
        sess.run(optimizer, {X: mean_normalize(df['X']), Y: df['Y']})
        
        loss = sess.run(loss_function, {X: mean_normalize(df['X']), Y: df['Y']})
        theta0_tmp = sess.run(theta0)
        theta1_tmp = sess.run(theta1)
        theta2_tmp = sess.run(theta2)
        
        if(step % 100 == 0):
            print("epoch: {0}, loss = {1}, theta0 = {2}, theta1 = {3}, theta2 = {4}".format(step, loss, theta0_tmp, theta1_tmp, theta2_tmp))
            
        summary = sess.run(loss_summary, {X: mean_normalize(df['X']), Y: df['Y']})
        writer.add_summary(summary, step)
        
    plt.scatter(df['X'], df['Y'], label='X')
    plt.scatter(df['X'], sess.run(y_pred, {X: mean_normalize(df['X'])}), c='r', label='Y')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.show()
```
