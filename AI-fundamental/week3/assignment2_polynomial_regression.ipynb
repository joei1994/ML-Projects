```python
import numpy as np
import tensorflow as tf
import matplotlib 
from matplotlib import pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('./data/fire_theft.xls')
```

```python
df.head()
```

```python
n_samples = len(df)
X = tf.placeholder(dtype='float32', name='X')
Y = tf.placeholder(dtype='float32', name='Y')

theta0 = tf.Variable(0.0, name='theta0')
theta1 = tf.Variable(0.0, name='theta1')
theta2 = tf.Variable(0.0, name='theta2')

y_pred = tf.multiply(theta2,(tf.pow(X, 2))) + tf.multiply(theta1, X) + theta0
mse = tf.multiply((1.0/ (2.0*n_samples)), tf.reduce_sum(tf.pow((y_pred - Y), 2)))
mse_summary = tf.summary.scalar(name='mse_summary', tensor=mse)

optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.00001).minimize(mse)

feed_dict = {X: df['X'],Y: df['Y']}
with tf.Session() as sess:
    writer = tf.summary.FileWriter('./graphs/assignment_polynomial/mse_summary')
    
    sess.run(tf.global_variables_initializer())
    
    for step in range(100000):
        sess.run(optimizer, feed_dict)
        
        mse_tmp = sess.run(mse, feed_dict)
        theta0_tmp = sess.run(theta0)
        theta1_tmp = sess.run(theta1)
        theta2_tmp = sess.run(theta2)
        if(step % 100 == 0):
            print("epoch:{0}, mse = {1}, theta0 = {2}, theta1 = {2}, theta2 = {3}".format(step,mse_tmp, theta0_tmp, theta1_tmp, theta2_tmp))
            
        summary = sess.run(mse_summary, feed_dict)
        writer.add_summary(summary, step)
    
    plt.scatter(df['X'], df['Y'], alpha=0.4, label='actual')
    plt.plot(df['X'], sess.run(y_pred, feed_dict), c='r', alpha=0.4, label='predictions')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.show()
```
