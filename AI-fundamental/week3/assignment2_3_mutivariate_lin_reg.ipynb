```python
import tensorflow as tf
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('./data/boston.csv')
```

```python
df.head()
```

```python
X_boston = df.drop(columns=['MEDV'], axis=1)
X_boston['B'] = 1
Y_boston = df['MEDV']
```

```python
n_samples = X_boston.shape[0]
X = tf.placeholder(dtype='float32', name='X', shape=[X_boston.shape[0], X_boston.shape[1]])
Y = tf.placeholder(dtype='float32', name='Y')

thetas = tf.Variable(tf.zeros(shape=[X_boston.shape[1], 1]), 'thetas')

y_pred = tf.matmul(X, thetas)
loss_function = tf.divide(tf.reduce_sum(tf.pow((y_pred - Y), 2)), (2.0 * n_samples))

optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.00).minimize(loss_function)
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for step in range(30000):
        sess.run(optimizer, {X: X_boston, Y: Y_boston})
        print("loss = {0}".format(sess.run(loss_function, {X: X_boston, Y: Y_boston})))
```
