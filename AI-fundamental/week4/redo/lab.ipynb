```python
plt.figure(figsize=(6, 6))
sns.scatterplot(data1[:, 0], data1[:, 1])
sns.scatterplot(data2[:, 0], data2[:, 1])
plt.show()
```

```python
data1 = np.append(data1, np.zeros(shape=(data1.shape[0], 1)), axis=1)
data2 = np.append(data2, np.ones(shape=(data2.shape[0], 1)), axis=1)

data = np.row_stack([data1, data2])
X_train = data[:, :2]
y_train = data[:, 2]
n_samples = X_train.shape[0]
n_features = X_train.shape[1]
```

```python
def calculate_decision_boundary(thetas, axis):
    lower, upper = axis.get_xlim()
    point1 = (lower, -(thetas[0] + thetas[1] * lower) / thetas[2])
    point2 = (upper, -(thetas[0] + thetas[1] * upper) / thetas[2])
    return point1, point2

def plot_data_with_decision_boundary(thetas):
    fig = plt.figure(figsize=(8, 5))
    ax = fig.add_subplot(111)
    ax.scatter(data1[:, 0], data1[:, 1], label='Class1')
    ax.scatter(data2[:, 0], data2[: ,1], label='Class2')

    point1, point2 = calculate_decision_boundary(thetas, ax)
    ax.plot([point1[0], point2[0]], [point1[1], point2[1]], label='decision boundary', c='r')
    plt.legend()
    plt.show()
```

```python
X = tf.placeholder(name='X', dtype=tf.float32, shape=[n_samples, n_features])
Y = tf.placeholder(name='Y', dtype=tf.float32, shape=[n_samples, 1])

thetas = tf.Variable(np.zeros(shape=(n_features, 1), dtype=np.float32), name='thetas')
theta0 = tf.Variable(0.0, name='theta0')

z = tf.matmul(X, thetas) + theta0
sigmoid = tf.div(1.0, 1.0 + tf.exp(-z))

cost_function = tf.reduce_mean(-(Y * tf.log(sigmoid) + (1 - Y) * tf.log(1-sigmoid)))

optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.03).minimize(cost_function)

feed_dict = {
    X : data[:, :2],
    Y : data[:, 2].reshape(-1, 1)
}
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for step in range(100000):
        sess.run(optimizer, feed_dict)
        
        if(step % 1 == 0):
            tts = sess.run(thetas)
            tt0 = sess.run(theta0)
            cost = sess.run(cost_function, feed_dict)
            print(f"step:{step}, cost:{cost:.3f}, theta0:{tt0:.3f}, theta1:{tts[0][0]:.3f}, theta2:{tts[1][0]:.3f}")
    
    tts = sess.run(thetas)
    tt0 = sess.run(theta0)
    plot_data_with_decision_boundary([tt0] + [tts[0][0], tts[1][0]])
```

### Problem 1.3 (Bonus: Implement LogistenRegression from scratch) 

```python
X_train = np.concatenate([np.ones(shape=(n_samples, 1)), X_train], axis=1)
```

```python
thetas = np.array([0.0, 0.0, 0.0])
converged = False
best_log_likelihood = -1e+6
accuracy = 0
epoch = 0
learning_rate = 0.03
```

```python
def z(x, thetas):
    return np.dot(x, thetas)

def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))

while not converged:
    random_indices = np.random.permutation(n_samples)
    for index in random_indices:
        x = X_train[index]
        y = y_train[index]
        
        p = sigmoid(z(x, thetas))
        
        thetas = thetas + learning_rate * (y - p) * x
        
    y_probs = sigmoid(z(X_train, thetas))
    y_preds = y_probs >= 0.5
    
    accuracy = sum(y_preds == y_train) / n_samples
    log_likelihood = np.sum(np.multiply(y_train, np.log(y_probs)) + np.multiply((1 - y_train), np.log(1 - y_probs)))
    
    if log_likelihood < best_log_likelihood + 1e-6:
        converged = True
    if log_likelihood > best_log_likelihood:
        best_log_likelihood = log_likelihood
    epoch = epoch + 1
```
