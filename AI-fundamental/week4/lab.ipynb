### Problem1.2

```python
#Prepare data

#Stack X1, X2
df1 = pd.DataFrame(data=X1, columns=['x1', 'x2'])
df1['class'] = 0
df2 = pd.DataFrame(data=X2, columns=['x1', 'x2'])
df2['class'] = 1

df = pd.concat([df1, df2], axis=0)
X_data = df.drop('class', axis=1)
X_data['x0'] = 1.0
X_data = X_data.reindex(columns=['x0','x1', 'x2'])

Y_data = df['class']
X_data.head()
```

```python
def decision_boundary(thetas, axis):
    lower, upper = axis.get_xlim()
    point1 = (lower, -(thetas[0] + thetas[1] * lower) / thetas[2])
    point2 = (upper, -(thetas[0] + thetas[1] * upper) / thetas[2])
    return point1, point2
```

```python
X = tf.placeholder(dtype='float32', shape=[X_data.shape[0], 3], name='X')
Y = tf.placeholder(dtype='float32', shape=[X_data.shape[0], 1], name='Y')

thetas = tf.Variable(tf.zeros(dtype='float32', shape=[X_data.shape[1], 1]), name='thetas')

z = tf.matmul(X, thetas)
pos_prob = tf.div(1.0, 1.0 + tf.exp(-z))
cost_function = tf.reduce_mean(-(Y * tf.log(pos_prob)) - (1-Y) * tf.log(1 - pos_prob))

optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.03).minimize(cost_function)

feed_dict = {X: X_data.values, Y: Y_data.values.reshape(-1, 1)}
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for step in range(30000):
        _, cost = sess.run([optimizer, cost_function], feed_dict)
      
        if(step % 100 == 0):
            theta0 = sess.run(thetas[0][0])
            theta1 = sess.run(thetas[1][0])
            theta2 = sess.run(thetas[2][0])
            print("Epoch {0}, loss = {1:.5}, theta0={2:.4}, theta1={3:.4}, theta2={4:.4}".format(step, cost, theta0, theta1, theta2))
            
    fig = plt.figure(figsize=(8, 5))
    ax = fig.add_subplot(111)
    ax.scatter(X1[:, 0], X1[:, 1], label='Class1')
    ax.scatter(X2[:, 0], X2[: ,1], label='Class2')
    
    theta0 = sess.run(thetas[0][0])
    theta1 = sess.run(thetas[1][0])
    theta2 = sess.run(thetas[2][0])
    point1, point2 = decision_boundary([theta0, theta1, theta2], ax)
    ax.plot([point1[0], point2[0]], [point1[1], point2[1]], label='decision boundary')
    plt.legend()
    plt.show()
```
